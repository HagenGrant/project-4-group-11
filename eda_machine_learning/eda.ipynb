```python
import pandas as pd
import numpy as np
from datetime import datetime as dt
from countryinfo import CountryInfo
```

```python
#Bringing in the dataset
file = "money_dataset_raw.csv"
df = pd.read_csv(file, encoding="cp1252")
df.head()
```

```python
##Changing column names
df.columns = [i.lower().replace(" ", "_") for i in df.columns]
df.columns = [i.lower().replace("(", "") for i in df.columns]
df.columns = [i.lower().replace(")", "") for i in df.columns]
```

```python
df.country.value_counts()
```

```python
df.info()
```

```python
#Changing data types
df['date_of_transaction'] = pd.to_datetime(df['date_of_transaction'])
df.info()
```

```python
#Adding in new date columns


df['month'] = pd.DatetimeIndex(df['date_of_transaction']).month
df.info()
```

```python
df.to_csv("cleaned_dataset.csv", index=False)
```
